```python
def process_data(data,country='Bangladesh',window=3):
    conf_case = data
    conf_case_country = conf_case[conf_case['Country/Region']==country]
    final_dataset = conf_case_country.T[4:].sum(axis='columns').diff().rolling(window=window).mean()[40:]
    df_conf = pd.DataFrame(final_dataset,columns=['Total'])
    return df_conf
```
